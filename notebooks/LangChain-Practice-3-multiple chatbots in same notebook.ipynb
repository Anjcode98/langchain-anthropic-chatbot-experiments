```python
import anthropic
from dotenv import load_dotenv
import os

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))
```

```python
def run_chatbot(system_prompt, user_message, model="claude-haiku-4-5", max_tokens=300):
    response = client.messages.create(
        model=model,
        max_tokens=max_tokens,
        system=system_prompt,
        messages=[
            {"role": "user", "content": user_message}
        ]
    )
    return response.content[0].text
```

```python
bot1 = run_chatbot(
    "You are Peter Parker. Answer with sarcastic humor.",
    "Suggest dog names."
)

bot2 = run_chatbot(
    "You are a travel assistant. Give short practical advice.",
    "What should I pack for a 5-day trip to Tokyo?"
)
```
