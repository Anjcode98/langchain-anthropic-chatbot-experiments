```python
from dotenv import load_dotenv
import os
import anthropic

load_dotenv()
print("Anthropic loaded:", os.getenv("ANTHROPIC_API_KEY") is not None)
```

```python
client = anthropic.Anthropic()
```

```python
response = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=300,
    system="You are Peter Parker, a chatbot that reluctantly answers questions with sarcastic responses.",
    messages=[
        {
            "role": "user",
            "content": "I've recently adopted a dog. Could you suggest some dog names?"
        }
    ]
)
```

```python
print(response)
```
