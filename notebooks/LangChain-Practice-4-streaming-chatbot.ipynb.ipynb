# LangChain Practice 4 - Claude Chatbot with Temperature and Streaming

## Objective
Practice using Anthropic Claude API with:
- temperature
- streaming
- reusable chatbot function

## What I learned
- how temperature changes response style
- how to stream responses token by token
- how to structure a reusable Claude helper function

load environment

```python
from dotenv import load_dotenv
import os

load_dotenv()

print("Anthropic key loaded:", os.getenv("ANTHROPIC_API_KEY") is not None)
```

create client

```python
import anthropic

client = anthropic.Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))
```

normal response with temperature

```python
response = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=300,
    temperature=0.7,
    system="You are a helpful assistant that explains things clearly.",
    messages=[
        {
            "role": "user",
            "content": "Write a short interview confirmation email."
        }
    ]
)

print(response.content[0].text)
```

lower temperature example

```python
response_low_temp = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=300,
    temperature=0.2,
    system="You are a professional HR assistant. Give formal and concise responses.",
    messages=[
        {
            "role": "user",
            "content": "Write a short interview confirmation email."
        }
    ]
)

print(response_low_temp.content[0].text)
```

streaming example

```python
with client.messages.stream(
    model="claude-haiku-4-5",
    max_tokens=300,
    temperature=0.8,
    system="You are a friendly assistant.",
    messages=[
        {
            "role": "user",
            "content": "Explain what a chatbot is in simple words."
        }
    ]
) as stream:
    for text in stream.text_stream:
        print(text, end="", flush=True)
```

reusable chatbot function

```python
def ask_claude(
    user_message,
    system_prompt="You are a helpful assistant.",
    model="claude-haiku-4-5",
    max_tokens=300,
    temperature=0.7,
):
    response = client.messages.create(
        model=model,
        max_tokens=max_tokens,
        temperature=temperature,
        system=system_prompt,
        messages=[
            {
                "role": "user",
                "content": user_message
            }
        ]
    )
    return response.content[0].text
```

test the function

```python
reply = ask_claude(
    user_message="Suggest five names for a brown dog.",
    system_prompt="You are a witty assistant.",
    temperature=0.9,
)

print(reply)
```
