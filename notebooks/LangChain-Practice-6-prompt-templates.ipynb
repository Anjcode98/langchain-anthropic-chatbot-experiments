## Initialize chat model

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("claude-haiku-4-5")
```

## 1. PromptTemplate

PromptTemplate allows us to create reusable prompts with variables.

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    "Explain {topic} in simple terms."
)

prompt = template.format(topic="Artificial Intelligence")


print(prompt)
```

Run with model

```python
response = model.invoke(prompt)
print(response.content)
```

# PromptTemplate (Multiple Variables)

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    "Explain {topic} for a {audience}"
)

prompt = template.format(
    topic="Machine Learning",
    audience="10-year-ols child"
)

response = model.invoke(prompt)
print(response.content)
```

## 2. PromptValue

PromptValue is the structured format LangChain uses internally.
It converts prompts into messages that models understand.

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    "Explain {topic} for a {audience}"
)

prompt_value = template.invoke({
    "topic": "Neural Networks",
    "audience": "beginner"
})

print(prompt_value)
```

```python
response = model.invoke(prompt_value)
print(response.content)
```

## 3. ChatPromptTemplate

Used for structured conversations with:
- system messages
- human messages

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helphul teacher."),
    ("human", "Explain {topic} in simple terms.")
])

prompt_value = chat_template.invoke({
    "topic": "Deep Learning"
})

response = model.invoke(prompt_value)
print(response.content)
```

## 4. Few-Shot Prompting

Few-shot prompting gives examples so the model learns the pattern.

```python
from langchain_core.prompts import FewShotPromptTemplate

examples = [
    {"input": "I love this product!", "output": "Positive"},
    {"input": "This is terrible.", "output": "Negative"},
]

example_template = PromptTemplate.from_template(
    "Input: {input}\nOutput: {output}"
)

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_template,
    suffix="Input: {user_input}\nOutput:",
    input_variables=["user_input"]
)

prompt = few_shot_prompt.format(user_input="The service was amazing!")

print(prompt)
```

```python
response = model.invoke(prompt)
print(response.content)
```

# Few-Shot Chat Example

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    ("human", "Translate 'Hello' to French."),
    ("ai", "Bonjour"),
    ("human", "Translate 'Good morning' to French.")
])

response = model.invoke(chat_template.invoke({}))
print(response.content)
```
