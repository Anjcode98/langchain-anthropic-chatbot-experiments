```python
from dotenv import load_dotenv
import os
import anthropic

load_dotenv()
print("Anthropic loaded:", os.getenv("ANTHROPIC_API_KEY") is not None)
```

```python
client = anthropic.Anthropic()
```

```python
professional_response = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=300,
    temperature=0.7,
    system="You are a professional HR assistant. Give clear and formal responses.",
    messages=[
        {   "role": "user",
            "content": "Write a short interview confirmation email."
        }
    ]    
)
```

```python
print(professional_response)
```
