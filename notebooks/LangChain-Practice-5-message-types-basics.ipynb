# LangChain Practice 5 - Understanding System, Human, and AI Messages

## Objective
Learn the difference between:
- System messages
- Human messages
- AI messages

## Why this matters
These message roles help control chatbot behavior and preserve conversation context.

Environment setup

```python
from dotenv import load_dotenv
import os

load_dotenv()
print("Anthropic key loaded:", os.getenv("ANTHROPIC_API_KEY") is not None)
```

Initialize chat model

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("claude-haiku-4-5")
```

## 1. System Message

A system message gives instructions to the model about how it should behave.

It sets the role, tone, and rules for the assistant.

Example:
- "You are a professional teacher."
- "You are a sarcastic assistant."
- "You always answer in short bullet points."

```python
from langchain_core.messages import SystemMessage, HumanMessage

message = [
    SystemMessage(content="You are a friendly math teacher. Explain answers simply."),
    HumanMessage(content="What is 12 multiplied by 8?")
]

response = model.invoke(message)

print(response.content)
```

## 2. Human Message

A human message is the user's input or question.

This is what the user says to the chatbot.

```python
message = [
    SystemMessage(content="You are a travel assistant. Answer briefly."),
    HumanMessage(content="What is the capital of Czech Republic?")
]

response = model.invoke(message)

print(response.content)
```

## 3. AI Message

An AI message represents a previous response from the assistant.

We use it when we want to continue a conversation and preserve context.

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

message = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="My favorite programming language is Python."),
    AIMessage(content="That's great. Python is popular for automation, AI, and web development."),
    HumanMessage(content="Can you remind me what I just said?")
]

response = model.invoke(message)

print(response.content)
```

## Here are a few good fresh examples.

Example 1: strict teacher

```python
messages = [
    SystemMessage(content="You are a strict science teacher. Keep answers short and serious."),
    HumanMessage(content="What is photosynthesis?")
]

response = model.invoke(messages)
print(response.content)
```

Example 2: funny assistant

```python
messages = [
    SystemMessage(content="You are a funny assistant who explains things with light humor."),
    HumanMessage(content="What is cloud computing?")
]

response = model.invoke(messages)
print(response.content)
```

Example 3: conversation memory using AI message

```python
messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="I live in Sri Lanka."),
    AIMessage(content="Nice! Sri Lanka is known for its natural beauty and rich culture."),
    HumanMessage(content="Where do I live?")
]

response = model.invoke(messages)
print(response.content)
```

Example 4: interview assistant

```python
messages = [
    SystemMessage(content="You are an interview coach. Give clear and professional answers."),
    HumanMessage(content="How should I introduce myself in a job interview?")
]

response = model.invoke(messages)
print(response.content)
```

## 4. Exercise

Try changing only the system message and observe how the same human message gets different responses.

```python
question = "Explain artificial intelligence."
```

```python
messages_1 = [
    SystemMessage(content="You are a university lecturer."),
    HumanMessage(content=question)
]

response_1 = model.invoke(messages_1)
print("Lecturer style:\n", response_1.content)
```

```python
messages_2 = [
    SystemMessage(content="You are explaining to a 10-year-old child."),
    HumanMessage(content=question)
]

response_2 = model.invoke(messages_2)
print("Child-friendly style:\n", response_2.content)
```

## Summary

- SystemMessage sets behavior and instructions
- HumanMessage represents the user's input
- AIMessage represents the assistant's previous reply
- Combining them helps create structured multi-turn conversations
